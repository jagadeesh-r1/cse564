```python
# Define your range
ra_min, ra_max = 275, 325
dec_min, dec_max = 25, 75

# Filter the data
df_filtered = df[(df['ra'] >= ra_min) & (df['ra'] <= ra_max) & (df['dec'] >= dec_min) & (df['dec'] <= dec_max)]
df_stars_filtered = df_stars[(df_stars['RA [deg]'] >= ra_min) & (df_stars['RA [deg]'] <= ra_max) & (df_stars['Dec [deg]'] >= dec_min) & (df_stars['Dec [deg]'] <= dec_max)]

# Plot the filtered data
plt.figure(figsize=(20, 10))
plt.scatter(df_filtered['ra'], df_filtered['dec'], s=10, c='blue', label='Exoplanets')
plt.scatter(df_stars_filtered['RA [deg]'], df_stars_filtered['Dec [deg]'], s=10, c='red', label='Stars')

plt.gca().invert_xaxis()  # RA increases from right to left
plt.xlabel('Right Ascension [deg]')
plt.ylabel('Declination [deg]')
plt.title('ExoPlanet and Star Map')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Prepare the data
planets_data = data[['ra', 'dec']].values

# Perform DBSCAN on planets data
db_planets = DBSCAN(eps=8, min_samples=30).fit(planets_data)
labels_planets = db_planets.labels_



# Add the cluster labels to the dataframes
data['cluster'] = labels_planets
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the 'ra' and 'dec' columns, colored by the 'cluster' column
plt.figure(figsize=(10, 8))
plt.scatter(data['ra'], data['dec'], c=data['cluster'], cmap='viridis')

# Add labels and title
plt.xlabel('ra')
plt.ylabel('dec')
plt.title('DBSCAN Clustering of Exoplanets')

# Show the plot
plt.show()
```

```python
numerical_cols = df_planets.select_dtypes(include=[np.number])
print(numerical_cols)
```

```python
correlation_matrix = numerical_cols.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs = corr_pairs[corr_pairs != 1]
sorted_pairs = corr_pairs.sort_values(kind="quicksort", ascending=False)
print(sorted_pairs)
```

```python
# Sort the indices of each pair
sorted_pairs = sorted_pairs.reset_index()
sorted_pairs[['level_0', 'level_1']] = np.sort(sorted_pairs[['level_0', 'level_1']].values, axis=1)

# Drop duplicate pairs
sorted_pairs = sorted_pairs.drop_duplicates(subset=['level_0', 'level_1'])

# Reset the index and print the sorted pairs
sorted_pairs = sorted_pairs.set_index(['level_0', 'level_1'])
# print(sorted_pairs)
# Extract unique column names
# Sort the indices of each pair
sorted_pairs = sorted_pairs.reset_index()

# Drop duplicate pairs
sorted_pairs = sorted_pairs.drop_duplicates()

# Extract unique column names
columns = pd.unique(sorted_pairs[['level_0', 'level_1']].values.ravel('K'))

print(columns)
```

```python
# Filter pairs based on correlation value
significant_pairs = sorted_pairs[(sorted_pairs[0] > 0.4) | (sorted_pairs[0] < -0.5)]

print(significant_pairs)
```
